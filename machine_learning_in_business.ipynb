###   –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

#### –§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–≤–æ–¥–∞ –≥—Ä–∞—Ñ–∏–∫–æ–≤:

```python
def func_graph(dataframe):
    fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(20,15))
    for i,j in enumerate(dataframe.drop("id",axis = 1)):
        axes[i][0].hist(dataframe[str(j)],bins = 100);
        axes[i][0].set_title(f"–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ '{str(j)}'")
        axes[i][0].set_xlabel(f"–ó–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ '{str(j)}'")#–∏–∑–º–µ–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ –æ—Å–∏ —Ö
        axes[i][0].set_ylabel("–ö–æ–ª-–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π ")#–∏–∑–º–µ–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ –æ—Å–∏ —É
        axes[i][1].boxplot( dataframe[str(j)])
        axes[i][1].set_ylabel(f"–ó–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ '{str(j)}'")#–∏–∑–º–µ–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ –æ—Å–∏ y
        axes[i][1].set_xlabel(f"{str(j)}")#–∏–∑–º–µ–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ –æ—Å–∏ y
        axes[i][1].set_title(f"–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞–∑–º–∞—Ö–∞ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ '{str(j)}'")
        plt.tight_layout()
    phik_overview = phik_matrix(dataframe.drop("id",axis = 1),interval_cols  = ['f0','f1','f2','product']) 
    plot_correlation_matrix(phik_overview.values,
    x_labels=phik_overview.columns,
    y_labels=phik_overview.index,
    vmin=0, vmax=1, color_map='Reds',
    title="–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π",
    fontsize_factor=1.5,
    figsize=(7, 6)) 
```

#### –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ geo_data_0 :

```python
func_graph(geo_data_0)
```

#### –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ geo_data_1:

## –û–±—É—á–µ–Ω–∏–µ –∏ –ø—Ä–æ–≤–µ—Ä–∫–∞ –º–æ–¥–µ–ª–∏

–°–æ–≥–ª–∞—Å–Ω–æ –¥–∞–Ω–Ω—ã–º –∏–∑ –¢–ó,–±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.–û—Ç–¥–µ–ª—å–Ω–æ —Å—Ç–æ–∏—Ç –∑–∞–º–µ—Ç–∏—Ç—å,—á—Ç–æ –≤—Å–µ –≤—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏–º–µ—é—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π –º–∞—Å—à—Ç–∞–±.–ù–µ—Å–º–æ—Ç—Ä—è –Ω–∞ —ç—Ç–æ,–ø—Ä–∏–º–µ–Ω–∏–º –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–º StandardScaler.

```python
def model_lr_geo_data(dataframe):
    X = dataframe.drop(['id',"product"], axis=1)
    y = dataframe['product']
    X_train, X_valid, y_train, y_valid = train_test_split(X, y,test_size=0.25,random_state=RANDOM_STATE)
    num_col_names = ["f0","f1","f2"]


    scaler = StandardScaler()
    # –æ–±—É—á–∏–º –∏ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏–∑ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ 
    X_train_scaled = scaler.fit_transform(X_train[num_col_names])
    # –≤—ã–ø–æ–ª–Ω–∏–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    X_valid_scaled = scaler.transform(X_valid[num_col_names])
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
    X_valid_scaled = pd.DataFrame(X_valid_scaled, columns=num_col_names)



    model_lr  = LinearRegression()
    # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    model_lr.fit(X_train_scaled, y_train)
    # –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö  –∏ —Å–æ—Ö—Ä–∞–Ω–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é predictions_0
    predictions = model_lr.predict(X_valid_scaled)




    predictions = pd.Series(predictions,name = "predict_product")
    df = pd.concat([y_valid.reset_index(drop = True),predictions],axis = 1)
    print(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {df['predict_product'].mean()}")
    print(f"–°—Ä–µ–¥–Ω–µ–µ  –∑–Ω–∞—á–µ–Ω–∏–µ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {df['product'].mean()}")
    # –ø—Ä–æ–≤–æ–¥–∏–º –æ—Ü–µ–Ω–∫—É —Ç–æ—á–Ω–æ—Å—Ç–∏
    rmse = (mean_squared_error(y_valid ,predictions))**0.5  
    print(f"–†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ RMSE –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å–∏–∏ –ø–æ –¥–∞–Ω–Ω—ã–º –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏  : {rmse}")

    
    
    residuals = df["product"] - df["predict_product"]
    # –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º ‚Äî –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —á–∞—Å—Ç–æ—Ç–Ω–æ—Å—Ç–∏ 
    # —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,7))
    axes[0].hist(residuals,bins = 35)
    axes[0].set_title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤')
    axes[0].set_xlabel('–û—Å—Ç–∞—Ç–∫–∏')
    axes[1].scatter(df["predict_product"],residuals)
    axes[1].set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏')
    axes[1].set_ylabel('–û—Å—Ç–∞—Ç–∫–∏')
    axes[1].set_title('–ê–Ω–∞–ª–∏–∑ –¥–∏—Å–ø–µ—Ä—Å–∏–∏')
    plt.show()
 

    return df,rmse
```

###  –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è  –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ geo_data_0 :

```python
df_0,rmse_0 = model_lr_geo_data(geo_data_0)
```

## –†–∞—Å—á—ë—Ç –ø—Ä–∏–±—ã–ª–∏ –∏ —Ä–∏—Å–∫–æ–≤ 

### –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–∏–±—ã–ª–∏

```python
def profit_calc(df):
    df_200 = df.sort_values(by='predict_product', ascending=False ).head(BEST_OBJECTS_NUM)
    return (df_200['product'].sum() * REVENUE_PER_UNIT) - TOTAL_BUDGET
```

### –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–µ—Ç–∞ –¥–æ—Ö–æ–¥–æ–≤ –∏ —É–±—ã—Ç–∫–æ–≤

```python
def profit_and_loss(df):
    state = np.random.RandomState(9)
    profit_bootstrep = []
    for i in range(0, 1000):
        df_sample = df.sample(
            n = OBJECTS_UNDER_STUDY_NUM, 
            replace = True, 
            random_state = state
        )
        profit_bootstrep.append(profit_calc(df_sample))
    profit_bootstrep = pd.Series(profit_bootstrep)
    # –°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥
    mean_profit = profit_bootstrep.mean()
    # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É–±—ã—Ç–∫–∞
    confidence_interval = st.t.interval(
    0.95, len(profit_bootstrep)-1,profit_bootstrep.mean() ,profit_bootstrep.sem()) 
    print(f'95% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –æ—Ç {confidence_interval[0]} —Ç—ã—Å.—Ä—É–±. –¥–æ {confidence_interval[1]}—Ç—ã—Å.—Ä—É–±.')
    probability_of_loss = profit_bootstrep[profit_bootstrep.values < 0].count()
    print('–°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥ –ø–æ —Ä–µ–≥–∏–æ–Ω—É   :', mean_profit)
    print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É–±—ã—Ç–∫–∞: {probability_of_loss/1000*100}%')
    profit_bootstrep.hist(bins = 50)
    
    lower_profit = profit_bootstrep.quantile(.025)
    # –í–µ—Ä—Ö–Ω–∏–π –∫–≤–∞–Ω—Ç–∏–ª—å 97.5% –æ—Ç 95% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
    upper_profit = profit_bootstrep.quantile(.975)
    print(lower_profit,upper_profit)
    return mean_profit, probability_of_loss
```

```python
mean_profit, probabilit—É_of_loss= profit_and_loss(df_0)
```
